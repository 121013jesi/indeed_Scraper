```python
from apify_client import ApifyClient
import csv

# Initialize the ApifyClient with your API token
client = ApifyClient("apify_api_AwyUzYNtefupFw6dfeZqnMcjzeNfTJ1UbACP")

# Prepare the Actor input
run_input = {
    "position": "web developer",
    "country": "US",
    "location": "San Francisco",
    "maxItems": 50,
    "parseCompanyDetails": False,
    "saveOnlyUniqueItems": True,
    "followApplyRedirects": False,
}

# Run the Actor and wait for it to finish
run = client.actor("hMvNSpz3JnHgl5jkh").call(run_input=run_input)

# Fetch items from the run's dataset
items = list(client.dataset(run["defaultDatasetId"]).iterate_items())

# Define the CSV file name
csv_file_name = 'output.csv'
                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                               
# Write the items to a CSV file
with open(csv_file_name, mode='w', newline='', encoding='utf-8') as file:
    if items:  # Check if there are items to write
        writer = csv.DictWriter(file, fieldnames=items[0].keys())
        writer.writeheader()  # Write the header
        writer.writerows(items)  # Write the data rows

print(f"Data saved to {csv_file_name}")
```
